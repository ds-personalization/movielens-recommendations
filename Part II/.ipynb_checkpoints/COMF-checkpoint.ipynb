```python
from lib import MEMF
from lib import MF
from lib import utils
from lib import preprocessing as prepro
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
ratings = pd.read_csv('../ml-latest/ratings_small.csv')
```

```python
ratings.head()
```

```python
hashed_ratings = utils.hash_data(ratings)
```

```python
t, reduced_ratings = prepro.separate_items_with_few_ratings(1, hashed_ratings)
```

```python
X_train, X_test,_,_ = train_test_split(reduced_ratings, 
                                       reduced_ratings['item'], 
                                       test_size = 0.15, 
                                       stratify = reduced_ratings['item'],
                                       random_state = 596)
```

```python
X_train = pd.concat([t, X_train])
```

```python
X_test[-X_test['user'].isin(X_train['user'].unique())]
```

```python
X_train = pd.concat([pd.DataFrame({'old_user' : [7026], 'old_item' : [590],
                                   'user': [254], 'item' : [24], 
                                   'rating' : [4.0]}), X_train], 
                    ignore_index = True)
X_test = X_test[-(np.logical_and(X_test['user'] == 254, X_test['item'] == 24))]
```

```python
y_train = X_train['rating']
y_test = X_test['rating']
X_train = X_train.drop('rating', axis = 1)
X_test = X_test.drop('rating', axis = 1)
```

```python
X_train.head()
```

```python
movie_file = "../ml-latest/clean_movies.csv"
```

```python
model = MEMF.COMF(movie_file)
```

```python
model.define_clusters()
```

```python
model.compute_membership()
```

```python
model.fit(X_train, y_train.values)
```

```python
for m in model.models.keys():
    print(model.tables[m])
```

```python
from tqdm import tqdm
```

```python
prediction = np.zeros(X_train['user'].size)

for index in tqdm(range(X_train['user'].size)):

    user = utils.get_true_user_at(index, X_train)
    item = utils.get_true_item_at(index, X_train)
    item_membership = model.membership[item]

    pred = 0
    # only consider the clusters than are relevant to our item
    for cluster in item_membership.keys():
        print("cluster " + str(cluster))
        # Getting back the dataframe restricted to the cluster
        table = model.tables[cluster].copy()
        print(table)

        m = model.models[cluster]
        weight = item_membership[cluster]
        print("user " + str(user))
        print("item "+ str(item))
        sel_user = table.loc[table['old_user'] ==user]
        print(sel_user)
        sel_item= table.loc[table['old_item'] == item]
        print(sel_item)
        selection = table.loc[(table['old_user'] == user) & (table['old_item'] ==item)]
        print(selection)
        temp = m.predict(selection)
        pred = pred + weight*temp[0]

    prediction[index] = pred
    break
```

```python
model.predict(X_train)
```

```python
model.score(y_train)
```
